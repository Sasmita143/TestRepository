## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`

### Box Plots


Plot a box plot of `Age.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Read the data into a DataFrame
QUERY = "SELECT * FROM master"
df = pd.read_sql_query(QUERY, conn)

# Convert 'Age' to numeric and drop NaNs
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df = df.dropna(subset=['Age'])

# Calculate Q1, Q3, and IQR
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

# Determine the bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

# Print the results
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers:\n{outliers}")

# Plot box plot of Age
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'], patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'),
            medianprops=dict(color='red'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Read the data into a DataFrame
QUERY = "SELECT * FROM master"
df = pd.read_sql_query(QUERY, conn)

# Convert 'Age' and 'WorkWeekHrs' to numeric and drop NaNs
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['WorkWeekHrs'] = pd.to_numeric(df['WorkWeekHrs'], errors='coerce')
df = df.dropna(subset=['Age', 'WorkWeekHrs'])

# Plot scatter plot of Age vs WorkWeekHrs
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], color='gray', alpha=0.6)
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(alpha=0.75)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Read the data into a DataFrame
QUERY = "SELECT * FROM master"
df = pd.read_sql_query(QUERY, conn)

# Convert 'WorkWeekHrs', 'CodeRevHrs', and 'Age' to numeric and drop NaNs
df['WorkWeekHrs'] = pd.to_numeric(df['WorkWeekHrs'], errors='coerce')
df['CodeRevHrs'] = pd.to_numeric(df['CodeRevHrs'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

# Create a bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age'] * 10, color='skyblue', alpha=0.6, edgecolors='black')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid(alpha=0.75)
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Read the data into a DataFrame
QUERY = "SELECT * FROM master"
df = pd.read_sql_query(QUERY, conn)

# Check a few rows to see what the data looks like
# print(df.head())

# Assuming you find a relevant column, let's say it's named 'MainBranch' for example:
# If 'MainBranch' contains database preferences:
database_counts = df['MainBranch'].str.split(';').explode().value_counts()

# Get the top 5 databases
top_databases = database_counts.head(5)

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Read the data into a DataFrame
QUERY = "SELECT * FROM master"
df = pd.read_sql_query(QUERY, conn)

# Filter data for the age group 30 to 35
age_group = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate the median for WorkWeekHrs and CodeRevHrs
median_work_week_hrs = age_group['WorkWeekHrs'].median()
median_code_rev_hrs = age_group['CodeRevHrs'].median()

# Prepare data for plotting
data = {
    'Metrics': ['Work Week Hours', 'Code Review Hours'],
    'Median Values': [median_work_week_hrs, median_code_rev_hrs]
}

# Create a DataFrame for plotting
median_df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(median_df['Metrics'], median_df['Median Values'], color=['darkblue', 'darkred'], label=median_df['Metrics'])
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.ylabel('Median Hours')
plt.xlabel('Metrics')
plt.grid(axis='y', alpha=0.75)
plt.legend()
plt.show()


```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Read the data into a DataFrame
QUERY = "SELECT * FROM master"
df = pd.read_sql_query(QUERY, conn)

# Filter data for the age group 45 to 60
age_group = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate the median ConvertedComp for each age
median_comp_by_age = age_group.groupby('Age')['ConvertedComp'].median()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-', color='mediumseagreen', linewidth=2, markersize=6)
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.xticks(median_comp_by_age.index)  # Set x-ticks to age values
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()


```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Read the data into a DataFrame
QUERY = "SELECT * FROM master"
df = pd.read_sql_query(QUERY, conn)

# Count occurrences of each MainBranch category
main_branch_counts = df['MainBranch'].value_counts()

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(main_branch_counts.index, main_branch_counts.values, color='coral')
plt.title('Distribution of Main Branches')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.grid(axis='x', alpha=0.75)
plt.tight_layout()
plt.show()
```
